```r
library(Seurat)
library(ggplot2)
library(ggsci)
library(RColorBrewer)
```

```r
load(file = "/data/Matthew/scrna/02.Clustering/crov7_alv_mincell5.Rdata")
DefaultAssay(seurat_obj) <- "MAGIC_RNA"
```

```r
id_path <- "/data/Matthew/scrna/02.Clustering/id_latest/"
celltype <- readLines(paste(id_path, "celltype.id",sep = "/"))
cellname <- readLines(paste(id_path, "celltype.name",sep = "/"))
pathway_id <- readLines(paste(id_path, "pathway_all_alkaloids.id",sep = "/"))
pathway_name <- readLines(paste(id_path, "pathway_all_alkaloids.name",sep = "/"))

p7 <- FeaturePlot(seurat_obj, features = celltype,
                  cols = c("lightgrey", brewer.pal(7, "YlOrRd")),
                  pt.size = 0.1, ncol = 5, order = T,
                  label = F, label.size = 3) & NoAxes() & NoLegend()
for (i in 1:length(celltype)) p7[[i]]$labels$title = cellname[i]
ggsave("celltype_FeaturePlot.png", p7, height = 22, width = 15)

p8 <- FeaturePlot(seurat_obj, features = pathway_id,
                  cols = c("lightgrey", brewer.pal(7, "YlOrRd")),
                  pt.size = 0.1, ncol = 5, order = T,
                  label = F, label.size = 3) & NoAxes() & NoLegend()
for (i in 1:length(pathway_id)) p8[[i]]$labels$title = pathway_name[i]
ggsave("pathway_all_FeaturePlot.png", p8, height = 35, width = 15)
```

```r
id_path <- "/data/Matthew/scrna/02.Clustering/id_latest/"
celltype_feature <- readLines(paste(id_path, "celltype_FeaturePlot.id",sep = "/"))
celltype_name <- readLines(paste(id_path, "celltype_FeaturePlot.name",sep = "/"))
pathway_feature <- readLines(paste(id_path, "pathway_FeaturePlot.id",sep = "/"))
pathway_names <- readLines(paste(id_path, "pathway_FeaturePlot.name",sep = "/"))
```

```r
p7 <- FeaturePlot(seurat_obj, 
                  features = celltype_feature, 
                  cols = c("lightgrey", brewer.pal(7, "YlOrRd")),
                  ncol = 3, order = T) & NoAxes() & NoLegend()
for (i in 1:length(celltype_feature)) p7[[i]]$labels$title = celltype_name[i]
ggsave("panels/celltype_FeaturePlot.png", p7, height = 7, width = 10, dpi = 600)

p8 <- FeaturePlot(seurat_obj, 
                  features = pathway_feature, 
                  cols = c("lightgrey", brewer.pal(7, "YlOrRd")), 
                  ncol = 3, order = T, label = T, label.size = 3) & NoAxes() & NoLegend()
for (i in 1:length(pathway_feature)) p8[[i]]$labels$title = pathway_names[i]
ggsave("panels/pathway_FeaturePlot.png", p8, height = 9, width = 15, dpi = 600)
```
